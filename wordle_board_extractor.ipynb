# Try to extract the board from a Wordle game

```python
# Load test image
from IPython.display import Image, display

# Show the image
display(Image(filename='images/benchmark.PNG'))
```

```python
from image_processing import get_wordle_grid_boxes
import cv2

image = cv2.imread('images/benchmark.PNG')

rows = get_wordle_grid_boxes(image)

if rows:
    # Draw contours on the original image in green
    for row in rows:
        for (cnt, x, y, w, h, cy) in row:
            cv2.drawContours(image, [cnt], -1, (0,255,0), 2)

        # Show the grayscale image
        cv2.imshow('Grayscale Image', image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
else:
    print(f"Failed to extract wordle board, found {len(rows)} rows instead of 6")
```
