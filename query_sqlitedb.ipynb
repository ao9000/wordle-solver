# Decode 1 row from guess_pattern

```python
# Import decoding functions
from preload_pattern_sqlite import decode_varints, decode_delta, decode_int_to_pattern


# Decode sample row from guess_pattern
guess_id, pattern_int, answer_blob = row
print(f"guess_id: {guess_id}")
print(f"pattern_int: {pattern_int}")
print(f"answer_blob: {answer_blob}")
```

```python
# Decode pattern_int to pattern tuple
pattern = decode_int_to_pattern(pattern_int)
print(f"pattern: {pattern}")
```

```python
# Decode answer_blob to answer_ids list
answer_ids_varints = decode_varints(answer_blob)
print(f"answer_ids_varints: {answer_ids_varints}")

# Decode answer_ids_varints to answer_ids using delta decoding
answer_ids = decode_delta(answer_ids_varints)
print(f"answer_ids: {answer_ids}")
```
